```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv("dt.csv")
my_players_set = ('R Laird', 'J Sicily', 'S Murray', 'J Finlayson', 
                   'I Cumming', 'N Coffield', 'A Pearce', 'A Naughton', 
                  'J McCrae', 'T Mitchell', 'N Fyfe', 'S Coniglio', 'P Dangerfield', 
                  'D Martin', 'Z Merrett', 'T Kelly', 'E Phillips', 'K Mutch', 'M Gawn', 
                  'S Martin', 'B Crossley', 'D Cameron', 'T McLean', 'I Heeney', 'S Menegola', 
                   'D Smith', 'B Fritsch', 'J Waterman', 'L Keeffe', 'C Rayner')
my_players_df = df.loc[df['player'].isin(my_players_set)]
scores_only = my_players_df.drop(['position', 'start_price', 'current_price', 'own', 'games','be'], axis=1)
scores_only_T = scores_only.set_index('player').T
```

```python
sns.boxplot(x="player", y="value", data=pd.melt(scores_only_T))
plt.show()
```

```python
my_players_df.to_csv("my_players.csv", columns=['player','round_1', 'round_2', 'round_3', 'round_4', 'round_5', 'round_6', 'round_7', 'round_8', 'round_9'], index=False)
```
