```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
%pylab inline
pylab.rcParams['figure.figsize'] = (25, 8)

df = pd.read_csv("dt.csv")
my_players_set = ('J McCrae', 'T Mitchell', 'N Fyfe', 'M Gawn', 'S Martin',
                  'S Coniglio', 'P Dangerfield', 'D Martin', 'Z Merrett', 'R Laird', 
                  'J Lloyd', 'J Sicily', 'A Brayshaw', 'T McLean', 'I Heeney', 
                  'S Menegola', 'D Smith', 'T Rockliff', 'E Phillips', 'B Crossley', 
                  'B Rice', 'B Fritsch', 'T Smith', 'J Waterman', 'L Keeffe', 'C Rayner')
my_players_df = df.loc[df['player'].isin(my_players_set)]
scores_only = my_players_df.drop(['position', 'start_price', 'current_price', 'own', 'games','be'], axis=1)
scores_only_T = scores_only.set_index('player').T
```

```python
sns.boxplot(x="player", y="value", data=pd.melt(scores_only_T))
sns.swarmplot(x="player", y="value", data=pd.melt(scores_only_T), color=".25")
plt.show()
```
