```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
%pylab inline
pylab.rcParams['figure.figsize'] = (25, 6)

df = pd.read_csv("dt.csv")
my_players_set = ('T Mitchell', 'M Gawn', 'B Grundy', 'M Crouch', 'D Beams', 'C Wingard',
                  'P Dangerfield', 'P Cripps', 'Z Merrett', 'R Laird', 'S Coniglio',
                  'J Lloyd', 'S Hurn', 'A Brayshaw',  'S Menegola', 'B Sier', 'J Dunkley',
                  'D Smith', 'T Rockliff', 'E Yeo', 'L Whitfield', 'S Ross', 'G Ablett')
my_players_df = df.loc[df['player'].isin(my_players_set)]
#scores_only = my_players_df.drop(['position', 'start_price', 'current_price', 'own', 'games','be'], axis=1)
scores_only_T = my_players_df.set_index('player').T
```

```python
sns.boxplot(x="player", y="value", data=pd.melt(scores_only_T))
sns.swarmplot(x="player", y="value", data=pd.melt(scores_only_T), color=".25")
plt.show()
```
